```python
datafile = sqlite3.connect("data.sqlite")
parts = extractor.get_parts(datafile)
parts
```

```python
data = extractor.get_partdata(datafile, "partie_Dictator_repetitions")
data.head()
```

```python
data.describe()
```

```python
playersA = data[data.DIC_role == 0]
playersA.head()
```

```python
fig, graph = plt.subplots(figsize=(8, 6))
graph.bar(range(len(playersA.joueur)), playersA.DIC_decision, width=0.2)
graph.set_xticklabels("")
graph.set_ylim(0, 10)
graph.set_ylabel("Amout sent")
graph.set_xlabel("Players")
graph.plot(playersA.DIC_decision.mean(), label="Average", color="red")
```
